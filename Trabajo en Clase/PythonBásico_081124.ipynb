```python
# prompt: Code to graph fourier spectrum

import numpy as np
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
# Assuming 'a' is a 1D array representing a time-domain signal
# Example:
a = np.sin(2 * np.pi * 5 * np.linspace(0, 1, 100)) + 0.5 * np.sin(2 * np.pi * 10 * np.linspace(0, 1, 100))


# Calculate the Fourier transform
spectrum = np.fft.fft(a)

# Calculate frequencies
frequencies = np.fft.fftfreq(len(a))

# Plot the spectrum
plt.figure(figsize=(10, 6))
plt.plot(frequencies, np.abs(spectrum))
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('Fourier Spectrum')
plt.grid(True)
plt.show()
```

```python
def deltadirac(x,x0):
  aux=np.zeros(len(x))
  cont=0
  for i in x:
    if round(i,7)==x0:
      aux[cont]=1
    else:
      aux[cont]=0
    cont+=1
  return np.array(aux)

t = np.arange(-10,10,0.1) #np.linspace (-10,10,100)
delta = deltadirac(t,0)

plt.figure(figsize=(8, 6))
plt.plot(t, delta, 'b.')
plt.xlabel('x')
plt.ylabel('δ(x)')
plt.title('Approximation of the Dirac Delta Function')
```

```python
class perro:
  def __init__(self,nombre,edad,raza,tamaño):
    self.nombre=nombre
    self.edad=edad
    self.raza=raza
    self.tamaño=tamaño
  def __call__(self):
    print('Hola soy un perro')
    self.ladrar()
    self.dormir()
    self.comer()
  def ladrar(self):
    print(self.nombre, 'Esta ladrando')
  def dormir(self):
    print(self.nombre, 'Esta durmiendo')
  def comer(self):
    print(self.nombre, 'Esta comiendo')
```

```python
Man = perro('NAN',13,'Labrador','Grande')
```

```python
Man.raza, Man.comer()
```

```python
class Empleado:
  def __init__(self,nombre,edad,puesto,salario):
    self.nombre=nombre
    self.edad=edad
    self.puesto=puesto
    self.salario=salario
  def obtener_salario(self):
    return self.salario

class AjusteSalario(Empleado):
  def __init__(self,nombre,edad,puesto,salario,aumento):
    super().__init__(nombre,edad,puesto,salario)
    self.aumento=aumento
  def obtener_salario_ajustado(self):
    return self.salario+self.aumento
```

```python
Em1 = Empleado('Juan',23,'Gerente',1000000)
Em1.obtener_salario()
```

```python
Em2 = AjusteSalario('Juan',23,'Gerente',1000000,200000)
Em2.obtener_salario_ajustado()
```

```python
class Producto:
  def __init__(self,nombre,precio):
    self.nombre=nombre
    self.precio=precio
  def obtener_precio(self):
    return self.precio

class ProductoDescuento(Producto):
  def __init__(self,nombre,precio,descuento):
    super().__init__(nombre,precio)
    self.descuento=descuento
  def obtener_precio_final(self):
    t= print(f'El producto {self.nombre} con precio base de {self.precio} tiene un descuento del {self.descuento}% y su precio final es de {self.precio-int(self.precio*self.descuento/100)}')
    return t
```

```python
P1 = Producto('Cuaderno',10000)
P1.obtener_precio()
```

```python
P2 = ProductoDescuento('Cuaderno',10000,20)
P2.obtener_precio_final()
```
